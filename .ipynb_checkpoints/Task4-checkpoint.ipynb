# Задача
Уравнение теплопроводности с начальными условиями первого рода.
Температура , например, на левой границе (x=0) линейно возрастает со временем.
На правой границе (x=b) в начальный момент времени температура равна температуре окружающей среды T0, 
Определить температуру во всем пространстве от x=0 до x=a за время t

# Функция решения поставленной задачи

```python
def solver(T_, a, h, t_, sigma, f, phi_1, phi_2):
    #T_ - температура окружающей среды
    #a - длина стержня
    #h - шаг по пространству
    #t_ - время
    #sigma - коэффициент теплопроводности
    #f - начальное условие
    #phi_1 - граничное условие на левом конце
    #phi_2 - граничное условие на правом конце
    N = int(a / h) # Количество шагов по координате
    dt = h * h/ sigma / 2 # сек шаг по времени
    M = int(t_ / dt) # Количество шагов по времени
    u = np.zeros((M, N)) # выделяем массив под вычисления
    # 
    def g(u, t, x):
        return 0 #Считаем что внешнего воздействия нет => g(...) = 0. В ином случае нужно поставить функцию.
    for j in range(N): # задаём начальное условие
        u[0][j] = f(j * h) 
    for k in range(M):
        u[k][0] = phi_1(k * dt) # температура на левой границе 
        u[k][N - 1] = phi_2(k * dt)  # температура на правой границе
    for i in range(u.shape[0]):
        print(u[i])
    plt.pcolormesh(u)
    
    for t in range(M - 1):
        for x in range(N - 1):  
            if x == 0:
                continue
            u[t + 1][x] = u[t][x] + sigma * dt / (h**2) * (u[t][x+1] - 2 * u[t][x] + u[t][x-1])
    return u # возвращаем решение в каждый шаг по времени
```

# Ввод начальных данных

```python
T_ = 20 #температура окружающей среды
def f(t): # начальное условие
    return T_
def phi_1(t): # граничное условие на левом конце
    return 5 * t + 20
def phi_2( t): # граничное условие на правом конце
    return T_ 
a = 1 # Длина стержня
h = 0.1 # шаг по пространству 
t_ = 1 # сек
sigma = 1 # коэффициент теплопроводности
```

# Решение задачи, вывод результатов, подвод итогов

```python
a = solver(T_, a, h, t_, sigma, f, phi_1, phi_2)
for i in range(a.shape[0]):
    print(a[i]) # выводим решение в каждый момент времени
#Видно что температура уменшьшается от левого конца к правому, что означает что программа работает корректно. 
#На правой границе у нас постоянная константа поскольку задачей не было задано как будет изменятся температура на правой границе.
#Если будет поставленно граничное условие на правом конце, то нужно не забыть его согласовать с начальным условием т.е. phi_2(0) == u[0][x] == T_ 
```
